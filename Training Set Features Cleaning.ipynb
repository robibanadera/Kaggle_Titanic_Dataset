# TRAIN DATA

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv('train.csv')
display(df)
display(df.info())
```

## Check and drop duplicates

```python
df = df.drop_duplicates(keep = 'first')
df
```

## Check for null values and replace or remove

```python
display(df.isnull().sum())
```

```python
null_cols = ['Age', 'Cabin', 'Embarked']

for i in null_cols:
    display(df[df[i].isnull() == True])
```

```python
df = df.drop(columns = ['Name', 'Cabin', 'Ticket'])
df.head()
```

```python
print('Old')
display(df.groupby('Embarked')['PassengerId'].count())

df['Embarked'].fillna(value='S', inplace=True)

print('New')
display(df.groupby('Embarked')['PassengerId'].count())
```

```python
print('Old')
display(df[df['Age'].isnull() == True])

age_mode = df['Age'].mode()
df['Age'].fillna(value=float(age_mode), inplace=True)

print('New')
display(df[df['Age'].isnull() == True])
```

```python
df.isnull().sum()
```

## Properly label the variables

```python
df_clean = df
```

```python
#Embarked
for place in df_clean['Embarked']:
    if place == 'S':
        df_clean['Embarked'] = df_clean['Embarked'].replace([place],'Southampton')
    elif place == 'C':
        df_clean['Embarked'] = df_clean['Embarked'].replace([place],'Cherbourg')
    elif place == 'Q':
        df_clean['Embarked'] = df_clean['Embarked'].replace([place],'Queenstown')

#Sex
df_clean['Sex'].replace({'male': 'Male', 'female': 'Female'}, inplace=True)

#Pclass
df_clean['Pclass'] = df_clean['Pclass'].astype(str)
df_clean['Pclass'].replace({'1':'1st Class', '2':'2nd Class', '3':'3rd Class'}, inplace=True)

df_clean.head()
```
